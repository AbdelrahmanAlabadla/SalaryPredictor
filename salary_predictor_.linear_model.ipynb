```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
df = pd.read_csv("Salary_Data[1].csv")

print(df.columns)
print(df.dtypes)
print(df.describe)
print(df.head)
print(df.isnull().sum())
```

```python
# Prepare the dataset

x=df['YearsExperience'].values.reshape(-1, 1)
y=df['Salary']

# Scale the feature
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)

# Split into train and test sets
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2)
model=LinearRegression()
model.fit(x_train,y_train)

# Predict on the test set
y_pred=model.predict(x_test)

# Evaluate the model
r2=r2_score(y_test,y_pred)
smse=np.sqrt(mean_squared_error(y_test,y_pred))
mae=mean_absolute_error(y_test,y_pred)

# User input prediction with validation
user_input = float(input("Enter the years of experience: "))
user_array=[[user_input]]
user_array_scaled=scaler.transform(user_array)
data_pred=model.predict(user_array_scaled)
print(f"Predicted salary: ${data_pred[0]:,.2f}")

# Print evaluation metrics with quotes
print(f'"R²": {r2:.4f}')
print(f'"RMSE": {rmse:.2f}')
print(f'"MAE": {mae:.2f}')
```
